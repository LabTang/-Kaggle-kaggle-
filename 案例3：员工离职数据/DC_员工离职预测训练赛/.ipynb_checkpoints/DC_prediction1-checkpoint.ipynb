```python
import pandas as pd #数据分析
import numpy as np #科学计算
from pandas import Series,DataFrame

data_train = pd.read_csv("./input/pfm_train.csv")
# data_train.head() #查看前5行数据
data_train
```

```python
data_train.info()
```

```python
data_train.describe()
```

```python
import matplotlib.pyplot as plt

# 没有以下两行，汉字将显示乱码
plt.rcParams['font.sans-serif']=['STKaiti'] #华文楷体
plt.rcParams['axes.unicode_minus']=False

fig = plt.figure()
fig.set_size_inches(16, 16)     #设置画布尺寸

plt.subplot2grid((2,4),(0,0))  
data_train.BusinessTravel.value_counts().plot(kind='bar')
plt.title(u"商旅频率")
plt.ylabel(u"人数")  

plt.subplot2grid((2,4),(0,1))  
data_train.Department.value_counts().plot(kind="bar")
plt.title(u"部门")
plt.ylabel(u"人数")

plt.subplot2grid((2,4),(0,2))  
data_train.EducationField.value_counts().plot(kind='bar')
plt.title(u"教育领域")
plt.ylabel(u"人数")  

plt.subplot2grid((2,4),(0,3))  
data_train.Gender.value_counts().plot(kind="bar")
plt.title(u"性别")
plt.ylabel(u"人数")

plt.subplot2grid((2,4),(1,0))
data_train.JobRole.value_counts().plot(kind='bar')
plt.title(u"工作角色")
plt.ylabel(u"人数") 

plt.subplot2grid((2,4),(1,1))
data_train.MaritalStatus.value_counts().plot(kind='bar')
plt.title(u"员工婚姻状况")
plt.ylabel(u"人数")

plt.subplot2grid((2,4),(1,2))
data_train.Over18.value_counts().plot(kind='bar')
plt.title(u"是否超过18")
plt.ylabel(u"人数")

plt.subplot2grid((2,4),(1,3))
data_train.OverTime.value_counts().plot(kind='bar')
plt.title(u"是否加班")
plt.ylabel(u"人数")

plt.tight_layout()
plt.show()
```
