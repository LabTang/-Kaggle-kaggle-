```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data_train = pd.read_csv('./input/train.csv')
data_train.shape
```

```python
data_test = pd.read_csv('./input/test.csv')
data_test.shape
```

```python
pd.set_option('display.max_columns',None)#显示所有列
pd.set_option('display.max_rows',None)#显示所有行
```

```python
data = pd.concat([data_train,data_test],axis = 0)
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
for i in data_train.columns:
    if data_train[i].dtype == 'float64':#float64 int64
        print(i + ':')
        print((data_train[data_train['Survived'] == 1.0][i].value_counts()/data_train[i].value_counts()).sort_values(ascending = False))
        print('-----------------------')
```

```python
for i in data_train.columns:
    if data_train[i].dtype == 'int64':#float64 int64
        print(i + ':')
        print((data_train[data_train['Survived'] == 1.0][i].value_counts()/data_train[i].value_counts()).sort_values(ascending = False))
        print('-----------------------')
```

```python
for i in data_train.columns:
    if data_train[i].dtype == 'O':
        print(i + ':')
        print((data_train[data_train['Survived'] == 1.0][i].value_counts()/data_train[i].value_counts()).sort_values(ascending=False))
        print('-----------------------')
```

```python
facet = sns.FacetGrid(data,hue = 'Survived' ,aspect=4)#aspect纵横比，是说每个小图的横轴长度和纵轴的比
facet.map(sns.kdeplot,'Age',shade = True)
facet.set(xlim=(0,data['Age'].max()))
facet.add_legend()
```
